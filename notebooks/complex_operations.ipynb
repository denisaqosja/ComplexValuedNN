```python
# verify that torch.nn.Conv2D performs the complex conjugate in the correlation

import torch
x = torch.tensor([1+2j, 3+4j])
y = x.conj()
print(f"{x = }")
print(f"{y = }")
print(y.is_conj())   # True  (flag is set)  
z = y.resolve_conj()
print(z.is_conj())   # False  
print(z)             # tensor([1-2j, 3-4j])   
```

```python
def complex_to_real(inp):
    inp_view_as_complex = torch.view_as_real(inp)
    dim_i = inp_view_as_complex.dim() - 1
    i_r = inp_view_as_complex.select(dim_i, 0)
    i_i = inp_view_as_complex.select(dim_i, 1)
    return (i_r, i_i)

print(complex_to_real(x.resolve_conj()))
```

```python
# Define the complex tensors
x = torch.tensor([1+2j, 3+4j])
h = torch.tensor([1+1j, 1-1j])

# Perform convolution
result = torch.conv1d(
    x.unsqueeze(0).unsqueeze(0),  # Add batch and channel dimensions
    h.flip(0).unsqueeze(0).unsqueeze(0),  # Flip for convolution and add dimensions
    padding=len(h)-1
)

# Perform convolution
result_noFlip = torch.conv1d(
    x.unsqueeze(0).unsqueeze(0),  # Add batch and channel dimensions
    h.unsqueeze(0).unsqueeze(0),  # Flip for convolution and add dimensions
    padding=len(h)-1
)

# Remove extra dimensions and trim to valid convolution size
print(result)
# result = result.squeeze()[:len(x) + len(h) - 1]
result_noFlip = result_noFlip.squeeze()[:len(x) + len(h) - 1]
print(result)
print(result_noFlip)
```
